```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Assuming store_df is already defined, and we will split the data for training and testing
store_train = store_df['Drivers_PersonsKilled_Number'][:-20]  # Use all but the last 20 for training
store_test = store_df['Drivers_PersonsKilled_Number'][-20:]  # Use the last 20 for testing

# Fit the ARIMA model on the training data
arima = ARIMA(store_train.astype(np.float64).to_numpy(), order=(1, 1, 1))
arima_model = arima.fit(start_params=[0.1, 0.1, 0.1])  # Example values, adjust as needed

# Use the model to make predictions on the test set
store_predict = arima_model.forecast(steps=len(store_test))

# Print the predictions
print("Predictions:\n", store_predict)

# Optionally, you can compare the predictions with the actual values
print("Actual Values:\n", store_test.values)
```

```python
import numpy as np

# Define the get_mape function
def get_mape(actual, forecast):
    """
    Calculate Mean Absolute Percentage Error (MAPE), ignoring zero actual values.
    """
    actual, forecast = np.array(actual), np.array(forecast)
    
    # Filter out zero actual values to avoid division by zero
    mask = actual != 0
    actual, forecast = actual[mask], forecast[mask]
    
    return np.mean(np.abs((actual - forecast) / actual)) * 100

# Assuming store_df and store_predict are already defined
# Using the last 20 values as the actual data for comparison
actual_values = store_df['Drivers_PersonsKilled_Number'][-20:]

# Calculate MAPE
mape = get_mape(actual_values, store_predict)
print(f"The ARIMA model gives forecast accuracy: {mape}%")

```
