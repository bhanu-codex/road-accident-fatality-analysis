```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# Load the Excel file
file_path ="C:/Users/Shanmukha Reddy/Desktop/FDS/Accidents Classified according to Non-Use of Safety Device ( Non-Wearing of Helmet) by Victims during the calendar year 2017-2022.xlsx"
wsb_df = pd.read_excel(file_path)

# Calculate a 12-month moving average for 'Drivers_PersonsKilled_Rank'
wsb_df['mavg_12'] = wsb_df['Drivers_PersonsKilled_Rank'].rolling(window=12).mean()

# Drop NaN values that result from the rolling calculation
wsb_df.dropna(subset=['mavg_12'], inplace=True)

# Calculate RMSE between 'Drivers_PersonsKilled_Rank' and its 12-month moving average from index 36 onward
rmse = np.sqrt(mean_squared_error(wsb_df['Drivers_PersonsKilled_Rank'][36:].values, wsb_df['mavg_12'][36:].values))
print("RMSE:", rmse)

# Compare the RMSE to a specific np.float64 value
comparison_value = np.float64(0.6357842046317105)
print("RMSE is equal to comparison value:", rmse == comparison_value)
```

```python
# Calculate Exponential Weighted Moving Average (EWMA) for 'Drivers_PersonsKilled_Rank'
wsb_df['ewm'] = wsb_df['Drivers_PersonsKilled_Rank'].ewm(alpha=0.2).mean()

# Display the DataFrame to see the new 'ewm' column
wsb_df[['Drivers_PersonsKilled_Rank', 'ewm']].head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the get_mape function to calculate Mean Absolute Percentage Error (MAPE)
def get_mape(actual, forecast):
    actual, forecast = np.array(actual), np.array(forecast)
    # Avoid division by zero by filtering out zero values
    mask = actual != 0
    actual, forecast = actual[mask], forecast[mask]
    return np.mean(np.abs((actual - forecast) / actual)) * 100

# Calculate MAPE between 'Drivers_PersonsKilled_Rank' and its Exponential Weighted Moving Average ('ewm')
mape_value = get_mape(wsb_df[['Drivers_PersonsKilled_Rank']][36:].values, wsb_df[['ewm']][36:].values)
print("MAPE:", mape_value)

# Compare the MAPE to the given value
comparison_value = np.float64(43266946650.46)
print("MAPE is approximately equal to comparison value:", np.isclose(mape_value, comparison_value))

# Plotting the actual vs. forecasted values using exponential moving average
plt.figure(figsize=(10, 4))
plt.xlabel('Months')
plt.ylabel('Quantity')
plt.plot(wsb_df['Drivers_PersonsKilled_Rank'][36:], label='Actual Drivers Persons Killed Rank')
plt.plot(wsb_df['ewm'][36:], '.', label='Exponential Moving Average (ewm)')
plt.legend()
plt.title('Actual vs Forecasted Drivers Persons Killed Rank using Exponential Moving Average')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np
import matplotlib.pyplot as plt

# Ensure the column 'Drivers_PersonsKilled_Rank' is used instead of 'Rank'
rank_series = wsb_df['Drivers_PersonsKilled_Rank'].dropna()  # Drop NaN values to avoid errors in decomposition

# Decomposing the time series with the additive model
ts_decompose = seasonal_decompose(rank_series, model='additive', period=12)

# Extracting the seasonal and trend components
wsb_df['seasonal'] = ts_decompose.seasonal
wsb_df['trend'] = ts_decompose.trend

# Plotting the decomposed time series components
ts_plot = ts_decompose.plot()
ts_plot.suptitle('Decomposition of Drivers Persons Killed Rank Time Series', fontsize=5)
plt.show()
```
